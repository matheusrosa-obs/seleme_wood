## Aplicando tratamento na coluna Categoria de Produto

```python
import unidecode
import pandas as pd

def normalize_category(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    text = unidecode.unidecode(text)
    text = text.strip()
    return text

df_itens["categoria_produto_norm"] = (
    df_itens["categoria_produto"]
        .astype(str)
        .apply(normalize_category)
)
```

```python
def keyword_grouping(cat):
    if not cat:
        return "outros"

    # PORTAS
    if "porta" in cat:
        return "portas"

    # MADEIRA
    if any(x in cat for x in ["wood", "madeira", "laminado"]):
        return "madeira"

    # MÓVEIS
    if any(x in cat for x in [
        "mobil",        # pega mobiliario, mobiliarios
        "movel",        # pega movel, moveis
        "sofa",
        "poltrona",
        "cadeira",
        "mesa",
        "buffet",
        "cristaleira",
        "aparador",
        "rack",
        "estante",
        "painel",
        "home"
    ]):
        return "moveis"

    # FERRAMENTAS
    if "ferramenta" in cat:
        return "ferramentas"

    return "outros"
```

```python
df_itens["categoria_produto_grupo"] = (
    df_itens["categoria_produto_norm"]
        .apply(keyword_grouping)
)
```

```python
import pandas as pd
import unidecode

def normalize_text(text):
    if pd.isna(text):
        return ""
    text = str(text).strip().lower()
    text = unidecode.unidecode(text)
    return text

df_itens["material_norm"] = df_itens["material"].apply(normalize_text)
```

```python
def materials_grouping(mat):
    if not mat:
        return "Não especificado"

    # "não informado"
    if any(x in mat for x in ["not specified", "not informed", "nao especificado", "nao informado", "unknown", "n/a"]):
        return "Não especificado"

    # Pinus (prioridade antes de madeira genérica, se você quiser manter separado)
    if any(x in mat for x in ["pinus", "pine"]):
        return "Pinus"

    # Madeira (wood/madeira)
    if any(x in mat for x in ["wood", "madeira", "Wood"]):
        return "Madeira"

    # fallback: só padroniza capitalização básica
    # (mantém "pet" -> "Pet", "aco baixo carbono..." -> "Aco baixo carbono...")
    return mat[:1].upper() + mat[1:] if mat else "Não especificado"
```

```python
df_itens
```

```python
df_itens["material_grupo"] = df_itens["material_norm"].apply(materials_grouping)
```

```python
import unicodedata

df_companies["nome_empresa_limpa"] = df_companies["nome_empresa"].str.rsplit("-", n=1).str[0].str.strip()
df_companies["cidade"] = df_companies["nome_empresa"].str.rsplit("-", n=1).str[-1].str.strip()

df_companies["cidade"] = (
    df_companies["nome_empresa"]
      .str.rsplit("-", n=1).str[-1].str.strip()
      .apply(lambda s: unicodedata.normalize("NFKD", s).encode("ascii", "ignore").decode("ascii"))
      .str.upper()
)
```

```python
geoloc_munic = geoloc_munic[['nm_mun', 'nm_uf', 'nu_latitude', 'nu_longitude']]
df_companies = df_companies.merge(geoloc_munic, left_on='cidade', right_on='nm_mun', how='left')
```
