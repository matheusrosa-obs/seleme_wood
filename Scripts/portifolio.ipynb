## Aplicando tratamento na coluna Categoria de Produto

```python
import unidecode
import pandas as pd

def normalize_category(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    text = unidecode.unidecode(text)
    text = text.strip()
    return text

df_itens["product_category_norm"] = (
    df_itens["product_category"]
        .astype(str)
        .apply(normalize_category)
)
```

```python
def keyword_grouping(cat):
    if not cat:
        return "outros"

    if "porta" in cat:
        return "portas"

    if any(x in cat for x in ["wood", "madeira", "laminado"]):
        return "madeira"

    if any(x in cat for x in ["painel", "rack", "estante", "home", "Mobiliário e utensílios", "Salas de Jantar",
                              "Cadeiras e Poltronas", "Sofás e Poltronas", "Mesa de Centro", "Aparadores", "Buffets e Cristaleiras",
                              "Móveis", "Mobiliário e utensílios", "Poltrona"]):
        return "moveis"

    if "ferramenta" in cat:
        return "ferramentas"

    return "outros"
```

```python
df_itens["product_category_grouped"] = (
    df_itens["product_category_norm"]
        .apply(keyword_grouping)
)
```
